#### Load dependencies

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import os # new!
```

```python
xs = np.array([0, 1, 2, 3, 4, 5, 6, 7.])
ys = np.array([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
```

points generate perturbing 
y = -0.5 + 2

```python
def gen(x):
    return (x * (-0.5)) + 2
```

```python
gen(22)
```

```python
model = Sequential()

model.add(Dense(1, input_dim=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.05))
```

```python
model.fit(xs, ys, epochs=200, verbose=1)
```

```python
model.predict(np.array([-10]))
```
